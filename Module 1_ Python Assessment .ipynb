##Mini Project: Banking System Using Python

###Account Login

```python
class BankAccount:
  def __init__(self,username,password):
    self.username=username
    self.password=password

  def login(self):
    attempts = 3

    while attempts >0:
      Username = input("Enter an username:")
      Password = input("Enter a password:")

      if Username==self.username and Password==self.password:
        print("Login Successful")
        break
      else:
        attempts-=1
        print(f"Wrong password.{attempts}attempts remaining")

Acc=BankAccount("maya","maya@000")
Acc.login()





```

###Amount Deposit

```python
class Deposit(BankAccount):
  def __init__(self,username,password,balance):
    super().__init__(username,password)
    self._account_type= "Saving"
    self.__balance=balance

  def Show_balance(self):
    print(f"Bank Balance: {self.__balance}")

  def deposit(self,amount):
    if amount>0:
      self.__balance +=amount


acc=Deposit("maya","maya@000",5000)
acc.Show_balance()
acc.deposit(1000)
acc.Show_balance()





```

###Amount Withdrawal

```python
class Withdrawal(Deposit):
  def __init__(self,username,password,balance):
    super().__init__(username,password,balance)
    self.__balance=balance

  def Show_balance(self):
    print(f"Bank Balance: {self.__balance}")

  def withdrawal(self,amount):
    if amount>0 and amount<=self.__balance:
        self.__balance -=amount


acc=Withdrawal("maya","maya@000",6000)
acc.Show_balance()
acc.withdrawal(500)
acc.Show_balance()

```
