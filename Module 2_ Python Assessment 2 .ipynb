```python
4.
team_salary=df.loc[0:457,['Team','Salary']]
team_salary
```

```python
team_salary.fillna(00000,inplace=True)
```

```python
team_salary.groupby('Team').sum().sort_values('Salary',ascending=True)

```

```python
plt.figure(figsize=(14,8))
sns.barplot(x=team_salary['Team'],y=team_salary['Salary'],hue=team_salary['Team'])
plt.title= 'Team vs Salary'
plt.xlabel('Team')
plt.ylabel('Salary')
plt.xticks(rotation=90)
```

####Observation:
The team 'Cleveland Cavaliers' have highest salary expenditure ,around 106988689 unit.And the team 'Philadelphia 76ers' have around 30992894 amount of salary.Which is less compaired to the salary received by other teams.
